### Carregando os dados das lojas e centralizando eles em um dicionário:

```python
arquivos_dados = ["loja_1.csv", "loja_2.csv","loja_3.csv", "loja_4.csv"]

dicionario_df = {}

for arquivo in arquivos_dados:
    key_dicionario = arquivo.split(".csv")[0]    
    loja = pd.read_csv(arquivo)
    dicionario_df[key_dicionario] = loja

dicionario_df.keys()
```

# 4. Produtos Mais e Menos Vendidos

```python
def busca_produtos_vendidos (agregacao="minimo"):
    dict_produtos_vendidos_lojas = {}
    lojas = []
    
    nome_produto = []
    count_produto = []
    
    for loja in dicionario_df.keys():
        _df = dicionario_df[loja].groupby('Produto')['Produto'].value_counts().to_frame()
        
        if agregacao == "minimo":
            count_produto.append(int(_df.min()))
            nome_produto.append(list(_df[_df['count'] == int(_df.min())].index))
            lojas.append(loja)
        
        elif agregacao == "maximo":
            count_produto.append(int(_df.max()))
            nome_produto.append(list(_df[_df['count'] == int(_df.max())].index))
            lojas.append(loja)
        else:
            print("Informe tipo de agreçação ('minimo' ou 'maximo')")
            break
    
    dict_produtos_vendidos_lojas['Loja'] = lojas
    dict_produtos_vendidos_lojas['Nome do Produto'] = nome_produto
    dict_produtos_vendidos_lojas['Produtos Vendidos'] = count_produto
    
    dict_produtos_vendidos_lojas = pd.DataFrame(dict_produtos_vendidos_lojas).set_index('Loja');
    return dict_produtos_vendidos_lojas
```

# 5. Frete Médio por Loja


```python
def custo_frete(lojas:list, metrica:str) -> pd.DataFrame:
    dict_ = {}
    lista_lojas = []
    frete = []
    
    for loja in lojas:
        if metrica == 'media':
            _metrica = dicionario_df[loja]['Frete'].mean()
            lista_lojas.append(loja)
            frete.append(_metrica)
        elif metrica == 'mediana':
            _metrica = dicionario_df[loja]['Frete'].median()
            lista_lojas.append(loja)
            frete.append(_metrica)
        else:
            print("Informe tipo de agregação ('media' ou 'mediana')")
            break
            
    dict_['Loja'] = lista_lojas
    dict_[f'Frete_{metrica}'] = frete
    return pd.DataFrame(dict_).set_index('Loja').sort_values(by=f'Frete_{metrica}', ascending=False)
```

```python
frete = custo_frete(lojas=dicionario_df.keys(), metrica='media')
frete
```

```python
frete = custo_frete(lojas=dicionario_df.keys(), metrica='mediana')
frete
```

# 6. Faturamento Total por Loja

```python
def faturamento_total(lojas:list) -> pd.DataFrame:
    dict_ = {}
    lista_lojas = []
    faturamento = []
    
    for loja in lojas:
        soma = dicionario_df[loja]['Preço'].sum()
        lista_lojas.append(loja)
        faturamento.append(soma)
    dict_['Loja'] = lista_lojas
    dict_['Faturamento'] = faturamento
    return pd.DataFrame(dict_).set_index('Loja').sort_values(by='Faturamento', ascending=False)
```

```python
faturamento = faturamento_total(lojas=dicionario_df.keys())
faturamento
```

### Evolução do faturamento mensal das lojas

```python
dicionario_faturamento_tempo = {}

for loja in dicionario_df.keys():
    dicionario_df[loja]['Data da Compra'] = pd.to_datetime(dicionario_df[loja]['Data da Compra'], format="%d/%m/%Y")
    faturamento_tempo = dicionario_df[loja].groupby(dicionario_df[loja]['Data da Compra'].dt.month)['Preço'].sum().to_frame()
    dicionario_faturamento_tempo[loja] = faturamento_tempo.sort_values(by='Data da Compra', ascending=True) #conjunto ordenado
```

```python
dicionario_faturamento_tempo
```

```python
fig, ax = plt.subplots(figsize=(12,6))
figura_faturamento_tempo = dicionario_faturamento_tempo.copy() #copia o dicionário para manipulá-lo na figura

while figura_faturamento_tempo.keys():
    try:
        loja = list(figura_faturamento_tempo.keys())[0] #obtem a primeira chave do dicionário
        ax.plot(figura_faturamento_tempo[loja], label=loja)
        ax.set_xlabel('Mês de Faturamento')
        ax.set_xticks(range(1, 13))
        ax.set_ylabel('Faturamento Mensal')
        ax.spines[['top','right']].set_visible(False)
        del figura_faturamento_tempo[loja] #após construção do gráfico, remove a chave já usada do dicionário temporário...
    except Exception as e:
        print(e)

ax.set_title("Faturamento mensal de lojas\nComparativo mensal")
plt.legend()
fig.tight_layout()
```
